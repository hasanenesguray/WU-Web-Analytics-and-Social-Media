**5. Detect the language**

```python
from langdetect import detect

def detect_language(comment):
    try:
        return detect(comment)
    except:
        return None

df['Language'] = df['Cleaned_Comment'].apply(detect_language)
df['Language'].value_counts()
```

```python
import matplotlib.pyplot as plt

average_lengths = df.groupby('Language')['Cleaned_Comment'].apply(lambda x: x.str.len().mean()).reset_index()
average_lengths.rename(columns={'Cleaned_Comment': 'Average_Length'}, inplace=True)

# Create a bigger bar chart
plt.figure(figsize=(10, 6))

# Plot the bar chart
bar_plot = plt.bar(average_lengths['Language'], average_lengths['Average_Length'])

# Add text annotations with the average lengths on the bars
for i, value in enumerate(average_lengths['Average_Length']):
    plt.annotate(f'{value:.2f}', xy=(i, value), ha='center', va='bottom')

plt.xlabel('Language')
plt.ylabel('Average Length')
plt.title('Average Comment Length by Language')

plt.show()
```

```python
import numpy as np

# Assuming 'df' is your DataFrame and 'Language' is the column containing language codes.

# Define the mapping of languages to replace
valid_languages = ["en", "af", "ro", "da", "no", "fr", "nl", "it", "tl", "et"]
replacement_language = "others"

# Use numpy.where() to apply the replacement condition
df['Language'] = np.where(df['Language'].isin(valid_languages), df['Language'], replacement_language)
df['Language'].value_counts()
```

```python
# Count the occurrences of each language
value_counts = df['Language'].value_counts()

# Sort the values in descending order
sorted_counts = value_counts.sort_values(ascending=True)

# Get the labels (unique values) and their respective counts
labels = sorted_counts.index.tolist()
counts = sorted_counts.tolist()

# Calculate percentages
total_count = sum(counts)
percentages = [(count / total_count) * 100 for count in counts]

# Create the bar chart with a bigger frame
plt.figure(figsize=(25, 8))  # Adjust the figure size as needed
bars = plt.barh(labels, counts, color=plt.cm.Paired.colors)

# Add count and percentage annotations on the bars
for bar, count, percentage in zip(bars, counts, percentages):
    plt.annotate(f'{count} ({percentage:.1f}%)', xy=(count, bar.get_y() + bar.get_height() / 2),
                 xytext=(5, 0), textcoords='offset points', va='center')

plt.xlabel('Count')
plt.ylabel('Language')
plt.title('Language Breakdown')
plt.tight_layout()  # To avoid label cutoffs
plt.show()
```

**9. Extracting noun phrases with TextBlob**

```python
from textblob import TextBlob
nltk.download('brown')

all_comments = ' '.join(df['Cleaned_Comment'])

# Create a TextBlob object
blob = TextBlob(all_comments)

# Extract noun phrases
noun_phrases = blob.noun_phrases

# Count the occurrences of each noun phrase using Counter
noun_phrase_counter = Counter(noun_phrases)

# Get the 10 most common noun phrases
most_common_noun_phrases = noun_phrase_counter.most_common(10)

# Print the results
print("10 most common noun phrases:")
for noun_phrase, frequency in most_common_noun_phrases:
    print(f"{noun_phrase}: {frequency}")
```

**10. Performing the LDA Topic Modelling**

```python
from gensim.models import LdaModel
from gensim.corpora import Dictionary
from nltk.tokenize import word_tokenize

nltk.download('punkt')

def tokenize_text(text):
    return word_tokenize(text)

df['Tokenized_Comment'] = df['Cleaned_Comment'].apply(tokenize_text)

# Create a dictionary from the tokenized words
dictionary = Dictionary(df['Tokenized_Comment'])

# Create a Bag-of-Words (BOW) representation for each document
bow_corpus = [dictionary.doc2bow(doc) for doc in df['Tokenized_Comment']]

# Train the LDA model
num_topics = 5  # You can set the number of topics as per your requirement
lda_model = LdaModel(corpus=bow_corpus, id2word=dictionary, num_topics=num_topics, passes=10)

# Print the topics and their top words
print("LDA Topics:")
for idx, topic in lda_model.print_topics():
    print(f"Topic {idx}: {topic}")
```

**11. Conduct a sentiment analysis with VADER model**

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()

# Function to get the sentiment score using VADER
def get_sentiment_score(text):
    return analyzer.polarity_scores(text)['compound']

# Apply the function to the "Sentiment_Comment" column
df['SentimentScore_VADER'] = df['Sentiment_Comment'].apply(get_sentiment_score)

# Categorize the sentiment based on the sentiment score
df['SentimentCategory_VADER'] = df['SentimentScore_VADER'].apply(lambda score: 'Positive' if score >= 0.05 else ('Negative' if score <= -0.05 else 'Neutral'))

# Print the DataFrame with sentiment scores and categories
df[['Sentiment_Comment', 'SentimentScore_VADER', 'SentimentCategory_VADER']]
```

**12. Conduct a sentiment analysis with TextBlob sentiment classifier**

```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import nltk
nltk.download('movie_reviews')

def get_sentiment_textblob(text):
    blob = TextBlob(text, analyzer=NaiveBayesAnalyzer())
    sentiment = blob.sentiment
    if sentiment.p_pos >= 0.6:
        return "Positive"
    elif sentiment.p_neg >= 0.6:
        return "Negative"
    return "Neutral"

df['SentimentCategory_Bayes'] = df['Sentiment_Comment'].apply(get_sentiment_textblob)

df[['Sentiment_Comment', 'SentimentCategory_Bayes']]
```

**13. Conduct a sentiment analysis with SocialSent**

```python
import pandas as pd

df_lex = pd.read_csv("2000.tsv", sep="\t", header=None)
df_lex.columns=["word","sentiment", "std.dev"]
df_lex
```

```python
mapping = {}
for row in df_lex[["word", "sentiment"]].to_dict("records"):
  mapping[row["word"]] = row["sentiment"]
mapping
```

```python
def sentiment_score(row):
  text = row["Sentiment_Comment"].lower()
  score = 0
  n = 0
  for word in text.split():
    if word in mapping:
      n += 1
      score += mapping[word]
  if n > 0:
    return score / n
  return 0

df["SentimentScore_SocialSent"] = df.apply(sentiment_score, axis=1)
df
```

```python
# Categorize the sentiment based on the sentiment score
df['SentimentCategory_SocialSent'] = df['SentimentScore_SocialSent'].apply(lambda score: 'Positive' if score >= 0.05 else ('Negative' if score <= -0.05 else 'Neutral'))
df
```

```python
bayes_counts = df['SentimentCategory_Bayes'].value_counts()
vader_counts = df['SentimentCategory_VADER'].value_counts()
social_sent_counts = df['SentimentCategory_SocialSent'].value_counts()

# Combine the sentiment categories from all three columns and sort them
sentiment_categories = sorted(set(bayes_counts.index) | set(vader_counts.index) | set(social_sent_counts.index))

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.25

x_bayes = range(len(sentiment_categories))
x_vader = [pos + bar_width for pos in x_bayes]
x_social_sent = [pos + bar_width * 2 for pos in x_bayes]

ax.bar(x_bayes, [bayes_counts.get(category, 0) for category in sentiment_categories], bar_width, label='Bayes')
ax.bar(x_vader, [vader_counts.get(category, 0) for category in sentiment_categories], bar_width, label='VADER')
ax.bar(x_social_sent, [social_sent_counts.get(category, 0) for category in sentiment_categories], bar_width, label='SocialSent')

ax.set_xticks([pos + bar_width for pos in x_bayes])
ax.set_xticklabels(sentiment_categories, rotation=45, ha='right')
ax.set_xlabel('Sentiment Category')
ax.set_ylabel('Count')
ax.set_title('Sentiment Analysis Comparison')
ax.legend()

plt.tight_layout()
plt.show()
```

**13. Sentiment analysis with machine learning**

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stopwords = stopwords.words('english')

def preprocess(row):
    text = row["Sentiment_Comment"].lower()
    keep = []
    for word in text.split():
        if word in stopwords:
            continue
        if word.startswith("@"):
            continue
        if word.startswith("http"):
            continue
        if word.startswith("#"):
            continue
        if word == "follow":
            continue
        if len(word) <= 1:
            continue
        keep.append(word)
    return ' '.join(keep)

df["cleaned_text"] = df.apply(preprocess, axis=1)
df
```

```python
from textblob.classifiers import NaiveBayesClassifier

train = df[['cleaned_text', 'SentimentCategory_VADER']][0:792]
test = df[['cleaned_text', 'SentimentCategory_VADER']][792:]

# Sample labeled data for sentiment analysis
training_data = [(row['cleaned_text'], row['SentimentCategory_VADER']) for _, row in train.iterrows()]

# Create a NaiveBayesClassifier and train it with the labeled data
classifier = NaiveBayesClassifier(training_data)

# Evaluate the classifier (optional)
testing_data = [(row['cleaned_text'], row['SentimentCategory_VADER']) for _, row in test.iterrows()]

accuracy = classifier.accuracy(testing_data)
print("Accuracy with VADER:", accuracy)

# Sentiment analysis on new text
new_text = "The book is fantastic!"
sentiment = classifier.classify(new_text)
print("Sentiment:", sentiment)  # Output: "positive"
```
