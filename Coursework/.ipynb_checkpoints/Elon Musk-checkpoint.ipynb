**1. Upload the dataset**

```python
import pandas as pd

# Specify the file path where the CSV file is located
csv_file_path = "reddit_comments.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)
df
```

```python
df.info()
```

**2. Adjust the type of the columns**

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df
```

**3. Apply TextBlob to correct comments**

```python
from textblob import TextBlob

# Function to apply TextBlob and get the corrected comment
def correct_comment(comment):
    blob = TextBlob(comment)
    corrected_comment = blob.correct()
    return str(corrected_comment)  # Convert the TextBlob object to a string

# Assuming 'df' is your DataFrame containing 'Comment_Body' column
df['Corrected_Comment'] = df['Comment_Body'].apply(correct_comment)
df
```

```python
df[df['Comment_Body'] != df['Corrected_Comment']]
```

**4. Detect the language**

```python
from langdetect import detect

def detect_language(comment):
    try:
        return detect(comment)
    except:
        return None

df['Language'] = df['Comment_Body'].apply(detect_language)
df['Language'].value_counts()
```

```python
import numpy as np

# Assuming 'df' is your DataFrame and 'Language' is the column containing language codes.

# Define the mapping of languages to replace
valid_languages = ["en", "fr", "it"]
replacement_language = "others"

# Use numpy.where() to apply the replacement condition
df['Language'] = np.where(df['Language'].isin(valid_languages), df['Language'], replacement_language)
df['Language'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'Column_Name' is the column for which you want to create the pie chart

# Calculate value counts for each unique value in the column
value_counts = df['Language'].value_counts()

# Get the labels (unique values) and their respective counts
labels = value_counts.index.tolist()
counts = value_counts.tolist()

# Calculate percentages
total_count = sum(counts)
percentages = [(count / total_count) * 100 for count in counts]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels, autopct='%.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is drawn as a circle.
plt.title('Pie Chart for Language')
plt.show()
```

**5. Remove the stopwords**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
```

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)

df['Without_Stopwords'] = df['Corrected_Comment'].apply(remove_stopwords)
```

```python
df[df['Corrected_Comment'] != df['Without_Stopwords']]
```

**6. Apply stemming**

```python
from nltk.stem import PorterStemmer

def apply_stemming(text):
    stemmer = PorterStemmer()
    words = word_tokenize(text)
    stemmed_words = [stemmer.stem(word) for word in words]
    return " ".join(stemmed_words)

df['Stemmed_Text'] = df['Without_Stopwords'].apply(apply_stemming)
df
```
